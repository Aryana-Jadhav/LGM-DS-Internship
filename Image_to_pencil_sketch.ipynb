```python
import sys
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
image = cv2.imread('/content/drive/MyDrive/LGM Internship/radhakrishna.jpg')
cv2_imshow(cv2.resize(image, (500, 500)))
cv2.waitKey(0)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(cv2.resize(gray_image, (500, 500)))
cv2.waitKey(0)
```

```python
inverted_image = 255 - gray_image
cv2_imshow(cv2.resize(inverted_image, (500, 500)))
cv2.waitKey()
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2_imshow(cv2.resize(blurred, (500, 500)))
cv2.waitKey()
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(cv2.resize(pencil_sketch, (500, 500)))
cv2.waitKey(0)
```
